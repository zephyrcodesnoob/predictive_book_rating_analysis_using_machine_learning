```python
# to represent the good books, best authors, best genres, best publishers as part of the dataset,
# as a binary attribute (1 = among the best, 0 = among the worst

# identification,

best_authors = list(updatedBooksDf['authors'].value_counts().sort_values(ascending = False).head(10).index)

best_publishers = list(updatedBooksDf['publisher'].value_counts().sort_values(ascending = False).head(10).index)

best_genres = list(updatedBooksDf['genre'].value_counts().sort_values(ascending = False).head(10).index)
```

```python
# using these values, we can create a feature of best_class,
# which will be 1 if the book is of best author, genre, publisher, will be 0 otherwise,
# in this way we introduce a feature that serves as a relationship between the 3 categorical values of (author, publisher and genre.)

finalBooksDf = updatedBooksDf.copy()
finalBooksDf.head()
```

```python
import numpy as np
finalBooksDf['best_author'] = np.where( np.isin( updatedBooksDf['authors'], best_authors) ==  1, 1, 0 ) # if in best then 1, else 0
```

```python
finalBooksDf.head()
```

```python
results = finalBooksDf.loc[finalBooksDf["best_author"] == 1]
results
```

```python
finalBooksDf['best_publisher'] = np.where( np.isin( updatedBooksDf['publisher'], best_publishers) == 1, 1, 0 ) # if in best then 1, else 0
```

```python
results = finalBooksDf.loc[finalBooksDf["best_publisher"] == 1]
results
```

```python
finalBooksDf['best_genre'] = np.where( np.isin( updatedBooksDf['genre'], best_genres) == 1, 1, 0 ) # if in best then 1, else 0
```

```python
results = finalBooksDf.loc[finalBooksDf["best_genre"] == 1]
results
```

```python
# a final feature best_class
def best_class(row):
  if row['best_author'] == 1 and row['best_publisher'] == 1 and row['best_genre'] == 1:
    return 1
  else:
    return 0
```

```python
finalBooksDf['best_class'] = finalBooksDf.apply(best_class, axis = 1) # axis = 1, i.e rows
```

```python
results = finalBooksDf.loc[finalBooksDf["best_class"] == 0]
results.head()
```

```python
# this function is used to get the columns that can be dropped effectively, depending on the correlation threshold value

def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
features_to_be_dropped = correlation(X_train, 0.7) # chosen threshold = 0.7, as observed from the above plot
features_to_be_dropped
```
